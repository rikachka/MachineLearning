<h4> WintersExponentialSmoothing

```python
# Winters Exponential Smoothing
# x <array Tx1>- time series, 
# h <scalar> - forecasting delay
# Params <dict> - dictionary with 
#    alpha <scalar in [0,1]> - smoothing parameter
#    delts <scalar in [0,1]> - seasonality smoothing parameter

def WintersExponentialSmoothing(x, h, Params):
    T = len(x)
    alpha = Params['alpha']
    delta = Params['delta']
    s = Params['seasonality_period']
    
    FORECAST = [np.NaN] * (T + h) ## здесь храните прогнозы
    
    a = 0 # инициализируйте значение
    b = np.zeros(s) # инициализируйте значение (это должен быть векотор длины s)
#     y = # инициализируйте значение
    
    for cntr in range(T):
#         print x[cntr]
        try:
            if not np.isnan(x[cntr]):
    #             if math.isnan(y):
    #                 y = # инициализируйте значение

                b[cntr % s] = delta * (x[cntr] - a) + (1 - delta) * b[cntr % s] # реализуйте формулу рекуррентного пересчёта
                a = alpha * (x[cntr] - b[cntr % s]) + (1 - alpha) * a # реализуйте формулу рекуррентного пересчёта

                #else do not nothing
            FORECAST[cntr + h] = a + b[(cntr + h) % s]
        except:
            FORECAST[cntr + h] = np.NaN
    return FORECAST
```

<h4> Подбор оптимального значения параметров \alpha и \delta на данных по departure 36

```python
def count_quality(ts, other_ts):
    ix = range(len(ts.index))
    QualityStr = qualityMedianAE(ts.sum(axis=1)[ix], other_ts.sum(axis=1)[ix])[0]
    return QualityStr
```

```python
def grid_search(ts, alphas, deltas):
    scores = []
    for alpha in alphas:
        scores_line = []
        for delta in deltas:
            w_frc_ts = get_w_frc_ts(ts, alpha, delta)
            score = count_quality(ts, w_frc_ts)
            scores_line.append(score)
        scores.append(scores_line)
    return scores
```

```python
def get_best_parametres(alphas, deltas, scores):
    scores_alpha = [min(scores_line) for scores_line in scores]
    alpha_index = scores_alpha.index(min(scores_alpha))
    alpha_best = alphas[alpha_index]

    scores_delta = scores[alpha_index]
    delta_index = scores_delta.index(min(scores_delta))
    delta_best = deltas[delta_index]

    score_best = scores[alpha_index][delta_index]

    print(alpha_best, delta_best, score_best)
    
    return alpha_best, delta_best
```

```python
ts = ts6
alphas = np.arange(0.1, 1.1, 0.2)
deltas = np.arange(0.1, 1.1, 0.2)
scores = grid_search(ts, alphas, deltas)
```

```python
alpha_best, delta_best = get_best_parametres(alphas, deltas, scores)
```

```python
def find_seasonable_series(ts, alpha, delta):
    w_frc_ts = get_w_frc_ts(ts, alpha, delta)
    es_frc_ts = get_es_frc_ts(ts)
    
    seasonable_cols = []
    
    for cntr in ts.columns:
        w_frc_ts_col = pd.DataFrame(index = w_frc_ts.index, columns = [cntr])
        w_frc_ts_col[cntr] = w_frc_ts[cntr]
        es_frc_ts_col = pd.DataFrame(index = es_frc_ts.index, columns = [cntr])
        es_frc_ts_col[cntr] = es_frc_ts[cntr]
        ts_col = pd.DataFrame(index = ts.index, columns = [cntr])
        ts_col[cntr] = ts[cntr]
        w_quality = count_quality_column(ts_col, w_frc_ts_col)
        es_quality = count_quality_column(ts_col, es_frc_ts_col)
        if w_quality > es_quality:
            seasonable_cols.append(cntr)
#             print cntr
    return seasonable_cols
```
